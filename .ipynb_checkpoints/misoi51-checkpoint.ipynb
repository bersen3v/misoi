```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/Users/ivanbersenev/Documents/code/python/VKDATAML/ikit.csv")
df = df.drop(['Unnamed: 0', 'home_town'], axis = 1)
df.columns = ["Имя", "Дата рождения","Город", "Пол", "Графовая группа", "Количество связей", "Центральность","Соеденительная центральность", "Близостная центральность", "Посредническая центральность", "Ранжированная центральность", "Идентификатор"]
df["Пол"] = np.where(df["Пол"] == 1, "Ж", "М")
for column in ["Центральность","Соеденительная центральность", "Близостная центральность", "Посредническая центральность", "Ранжированная центральность"]: 
    df[column] = round((df[column] - df[column].min()) / (df[column].max() - df[column].min()), 2)
# df['birthday'] = pd.to_datetime(df.bdate, errors="coerce")
# df['birthday_month'] = df.birthday.dt.month
# df['birthday_year'] = df.birthday.dt.year
```

```python
# df.head(5)
styles = [{
    'selector': 'th',
    'props': [('text-align', 'left')]
}]

categorical_columns = df.select_dtypes(['object']).columns
for col in categorical_columns:
    styles.append({'selector': f'[data-columns="{col}"]', 'props': [('text-align', 'left')]})

def align_text(x):
    if isinstance(x, str):
        return 'text-align: left'
    else:
        return 'text-align: right'


styled_df = df.style.set_table_styles(styles)
styled_df = df.style.format('{:.2f}').applymap(align_text, subset=pd.IndexSlice[:, :])

styled_df
```

```python
def formatting(val):
    color = 'red' if val < 100 else 'black'
    return f'color: {color}'

df.head(10).style.map(formatting, subset = ["Количество связей"])
```

```python
df.sort_values(by="Центральность", ascending=False).head(5)
```

```python
# Форматирование цвета ячеек по их величине
max_value = df['Количество связей'].max()
def formatting(val):
    ratio = val / max_value
    r = int( max(0, min(255, 255 * (1 - ratio))))
    g = int(max(0, min(255, 255 * ratio)))
    color = f'rgba({r},{g},0,1)'
    return f'color: {color}'

df.sort_values(by="Количество связей", ascending = False).head(10).style.map(formatting, subset = ["Количество связей"])
```
