```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("/Users/ivanbersenev/Documents/code/python/VKDATAML/ikit.csv")
```

```python
# получаем только те строки, пользователи которых находятся в пятерке городов
df_top_cities = df[df.city.apply(lambda x: x in top_cities)]
df_top_cities.head(5)
```

```python
#Линейная диаграмма зависимости eigenvector_centrality от degree
sns.lineplot(df.head(100), x = "eigenvector_centrality", y = "degree")
```

```python
#Столбчатая диаграмма зависимости eigenvector_centrality от degree
sns.barplot(df.head(100), x = "eigenvector_centrality", y = "degree")
```

```python
#Получаем даты рождения тех, у кого они указаны
df['birthday'] = pd.to_datetime(df.bdate, errors="coerce")
```

```python
# находим пять самых популярных городов среди подписчиков
top_cities = list(df.groupby("city").agg({"id": "count"}).sort_values(by = "id", ascending = False).head(5).index)
top_cities
```

```python
#Рисуем гистограмму по годам рождения с накоплением по полу
sns.histplot(df, x = "birthday", hue = "sex", multiple='layer')
```

```python
#гистограмма по пятерке самых популярных городов
df_top_cities.city.hist()
```

```python
# Строим график по городам с учетам пола, где 2 - м, 1 - ж
sns.histplot(data = df_top_cities, x = "city", hue = "sex")
```

```python
#Рисуем гистограмму по дням рождения
df['birthday_day'] = df.birthday.dt.day
sns.histplot(df, x = "birthday_day")
```

```python
#Рисуем гистограмму по месяцам рождения
df['birthday_month'] = df.birthday.dt.month
df['birthday_year'] = df.birthday.dt.year
sns.histplot(df, x = "birthday_month")
```

```python
# зависимость eigenvector_centrality от degree. Точечная диаграмма
sns.lmplot(
    data=df, x="eigenvector_centrality", y="degree" , hue = "sex"
)
```

```python
# зависимость closeness_centrality от degree. Точечная диаграмма
sns.lmplot(
    data=df, x="closeness_centrality", y="degree", hue = "sex"
)
```

```python
# зависимость betweenness_centrality от degree. Точечная диаграмма
sns.lmplot(
    data=df, x="betweenness_centrality", y="degree", hue = "sex"
)
```

```python
# зависимость closeness_centrality от eigenvector_centrality. Точечная диаграмма
sns.lmplot(
    data=df, x="closeness_centrality", y="eigenvector_centrality", hue = "sex"
)
```

```python
# Круговая диаграмма по городу подписчиков
data = df.groupby("city").agg({"id": "count"}).sort_values(by = "id", ascending = False).head(5) 
colors = sns.color_palette("bright")
plt.pie(list(data.id), labels=list(data.index), colors=colors, autopct="%0.0f%%")
```

```python
# Лепестковая диаграмма по количеству людей в разных community
categories = list(df.community.index)
num_categories = len(list(df.community))

angles = np.linspace(0, 2 * np.pi, num_categories, endpoint=False).tolist()
angles += angles[:1]

values = list(df.community)
values += values[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))

ax.fill(angles, values, color='skyblue', alpha=0.25)
ax.plot(angles, values, color='skyblue', linewidth=1)

plt.show()
```
