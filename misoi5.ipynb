```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/Users/ivanbersenev/Documents/code/python/VKDATAML/ikit.csv")
df = df.drop(['Unnamed: 0', 'home_town'], axis = 1)
df.columns = ["Имя", "Дата рождения","Город", "Пол", "Графовая группа", "Количество связей", "Центральность","Соеденительная центральность", "Близостная центральность", "Посредническая центральность", "Ранжированная центральность", "Идентификатор"]
df["Пол"] = np.where(df["Пол"] == 1, "Ж", "М")
for column in ["Центральность","Соеденительная центральность", "Близостная центральность", "Посредническая центральность", "Ранжированная центральность"]: 
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())     
# df['birthday'] = pd.to_datetime(df.bdate, errors="coerce")
# df['birthday_month'] = df.birthday.dt.month
# df['birthday_year'] = df.birthday.dt.year
```

```python
# Данные
# Имя - имя подписчика группы вконтакте Икит
# Дата рождения 
# Город 
# Пол
# Графовая группа - группа пользователя, посчитанная по алгоритму лувена на основе близости к другим точкам
# Количество связей - количество друзей пользователя, которые тоже подписаны на группу икита
# Центральность - центральность ноды, рассчитанная как degree деленое на максимальное degree
# Соеденительная центральность - коэффициент важности узла. Увеличивается пропорционально тому, сколько сообществ соединяет нода
# Близостная центральность - показывает, насколько близко узел расположен к остальным узлам сети
# Посредническая центральность - важность узла на основе важности его соседей
# Ранжированная центральность - важность узла, рассчитанная анологично тому, как рассчитывается важность страниц в google
# Идентификатор - id пользователя внутри Вконтакте
```

```python
df = pd.read_csv("/Users/ivanbersenev/Documents/code/python/VKDATAML/ikit.csv")
communities_sizes = df[["community", "name"]].groupby('community').count().reset_index().sort_values(by="name", ascending=False)
sizes = communities_sizes.groupby('name').count().reset_index()
sizes['groupNumber'] = communities_sizes.community
sizes['community'] = np.where(sizes['community'] == 1, sizes['name'], sizes['community'])
sizes['groupName'] = sizes.groupNumber.apply(lambda x: "люди, не находящиеся ни в одной группе" if x == 0 else f"группа {x}")
sizes = sizes.set_index('groupName')
sizes = sizes.drop('name', axis=1)
sizes = sizes[sizes.community > 100]
```

```python
axs = sizes.community.plot.pie(subplots=True, labels=None, autopct='%1.1f%%',pctdistance=1.1, fontsize=6 )
for ax in axs:
    ax.set_xlabel(None)
    ax.set_ylabel(None)
    
plt.legend(sizes.community.index, loc='lower left', bbox_to_anchor=(1, 0.1))
plt.title("Соотношение размеров графовых сообществ, образованных подписчиками ИКИТ (4600 чел.)")
```

```python
df = pd.read_csv("/Users/ivanbersenev/Documents/code/python/VKDATAML/ikit.csv")
top_cities = list(df.groupby("city").agg({"id": "count"}).sort_values(by = "id", ascending = False).head(5).index)
df_top_cities = df[df.city.apply(lambda x: x in top_cities)]
plot = df_top_cities.city.hist()
plt.title("Распределение подписчиков сообщества ИКИТ по городам")
plot.set_ylabel("Количество людей")
plot.set_xlabel("Название города")
new_labels = ['Красноярск', 'Москва', 'Санкт-Петербург', 'Железногорск', 'Улан-Уде']
plot.set_xticklabels(new_labels, rotation=45) 
```
